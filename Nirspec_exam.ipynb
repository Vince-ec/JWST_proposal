```python
from astropy.table import Table
from astropy.io import fits
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from matplotlib import gridspec
import matplotlib.patches as patches
from glob import glob
import seaborn as sea
import numpy as np
import pandas as pd
sea.set(style='white')
sea.set(style='ticks')
sea.set_style({'xtick.direction': 'in','xtick.top':True,'xtick.minor.visible': True,
               'ytick.direction': 'in','ytick.right': True,'ytick.minor.visible': True})
cmap = sea.cubehelix_palette(12, start=2, rot=.2, dark=0, light=1.2, as_cmap=True)

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
wv140,g140trans = np.load('g140m_f100lp_trans.npy') ; wv235,g235trans = np.load('g235m_f170lp_trans.npy') ; wv395,g395trans = np.load('g395m_f290lp_trans.npy')
```

```python
wv140,fnu140,err140 = np.load('g140_z5_spec.npy') ; wv235,fnu235,err235 = np.load('g235_z5_spec.npy') ; wv395,fnu395,err395 = np.load('g395_z5_spec.npy')
```

```python
g140trans /= max(g140trans)
g235trans /= max(g235trans)
g395trans /= max(g395trans)
```

```python
plt.figure(figsize=[12,6])
plt.plot(wv140,g140trans)
plt.plot(wv235,g235trans)
plt.plot(wv395,g395trans)
```

```python
adderr140 = np.random.normal(0,err140)
adderr235 = np.random.normal(0,err235)
adderr395= np.random.normal(0,err395)
```

```python
plt.figure(figsize=[20,6])
plt.plot(wv140,fnu140/g140trans)
plt.plot(wv235,fnu235/g235trans)
plt.plot(wv395,fnu395/g395trans)
plt.xlim(1,5.15)

plt.figure(figsize=[20,6])
plt.plot(wv140,fnu140/err140)
plt.plot(wv235,fnu235/err235)
plt.plot(wv395,fnu395/err395)
plt.xlim(1,5.15)

plt.figure(figsize=[20,6])
plt.errorbar(wv140,(fnu140+adderr140)/g140trans,err140/g140trans,fmt='o', ms=2,zorder=1)
plt.errorbar(wv235,(fnu235+adderr235)/g235trans,err235/g235trans,fmt='o', ms=2,zorder=1)
plt.errorbar(wv395,(fnu395+adderr395)/g395trans,err395/g395trans,fmt='o', ms=2,zorder=1)
plt.plot(wv140,fnu140/g140trans,'k',zorder=2)
plt.plot(wv235,fnu235/g235trans,'k',zorder=2)
plt.plot(wv395,fnu395/g395trans,'k',zorder=2)
plt.xlim(1,5.15)
plt.ylim(-4,20)
```
